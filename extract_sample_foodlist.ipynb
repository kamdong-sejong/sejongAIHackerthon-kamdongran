# foodlist의 식품군 별로 sample추출하여 저장

```python
import numpy as np
import pandas as pd
```

```python
# 최대 줄 수 설정
pd.set_option('display.max_rows', 5000)
# 최대 열 수 설정
pd.set_option('display.max_columns', 100)
# 표시할 가로의 길이
pd.set_option('display.width', 1000)
```

```python
raw= pd.read_excel('foodlist.xlsx')
```

```python
raw['식품군'].unique()
```

```python
isnt_del = (raw['식품군'] != '김치류')&(raw['식품군'] != '나물류')&(raw['식품군'] != '떡류')&(raw['식품군'] != '아이스크림류')&(raw['식품군'] != '음료류')&(raw['식품군'] != '장아찌류')&(raw['식품군'] != '포류')&(raw['식품군'] != '한과류')
```

```python
df=raw[isnt_del]
```

```python
df2 = df.drop('제조사',axis=1)
df3=df2.reset_index(drop=True)
```

```python
df3['식품군'].unique()
```

```python
new=df3
```

```python
new2=new.drop(index=new.index)
new2
```

```python
list=df3['식품군'].unique()
list
```

```python
for j in range(len(list)):
    for i in range(len(df3)):
        if (df3.iloc[i]['식품군']==list[j]):
            new2=new2.append(df3.loc[i:i+4])
            break;
new3=new2.reset_index(drop=True)
```

```python
# 최대 줄 수 설정
pd.set_option('display.max_rows', 1500)
# 최대 열 수 설정
pd.set_option('display.max_columns', 100)
# 표시할 가로의 길이
pd.set_option('display.width', 1000)
new3
```

```python
new3.to_excel('foodsample.xlsx',sheet_name='Sheet1')
```
