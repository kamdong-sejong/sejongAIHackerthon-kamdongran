```python
raw= pd.read_excel('foodsample.xlsx')
```

```python
raw['dine_number']=0
raw['name']=''
raw.rename(columns = {'식품명' : 'main_food'}, inplace = True)
raw.rename(columns = {'식품군' : 'food_type'}, inplace = True)
```

```python
df=raw.drop(raw.columns[[2,3,4,5,6,7,8,9,10]], axis='columns')
```

```python
df['longtitude']=0.0
df['latitude']=0.0
df['main_ingridient']=''
```

```python
df=df[['dine_number','name','main_food','main_ingridient','food_type','longtitude','latitude']]
df[['dine_number']]=df.index
df[['name']]=df.index
```

알러지 종류 : ['소','돼지','새우','밀','우유']

알러지 유발 재료가 포함된 음식에 알러지 정보를 입력한다.

```python
al=['소','돼지','새우','밀','우유']
```

```python
for i in range(len(df)):
    for j in range(len(al)):
        if(al[j] in df.loc[i,'main_food']):
            df.loc[i,'main_ingridient']=al[j]
            break;
```

각 식당의 위도와 경도를 건대와 어린이대공원 주변으로 임의로 설정한다.

longtitude(경도) 127.05~127.09

latitude(위도) 37.55~37.53

```python
for i in range(len(df)):
    long=uniform(127.05,127.09)
    lat=uniform(37.55,37.53)
    df.loc[i,'longtitude']=long
    df.loc[i,'latitude']=lat
```
